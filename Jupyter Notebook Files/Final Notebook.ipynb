```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
# File to Load 
clusters_data = "Map geographical clusters.csv"

# Read aquacultral farm data File and store into Pandas data frame
df = pd.read_csv("Map geographical clusters.csv")
df.head()
```

```python
df['Area_log'] = np.log(df['Area (ha)'])
df
```

```python

nitrogen = df.to_html()
print(nitrogen)
```

```python
locations = []

for index, rows in df.iterrows():
    locations.append((rows.Latitude_p, rows.Longitude_p))
    
locations
```

```python
# import the library
import folium

# Make an empty map

new_m = folium.Map(location=[35,14], tiles="OpenStreetMap", zoom_start=4.3, control_scale = True)

# Show the map
new_m
```

```python
for i in range(0,len(df)):
   folium.CircleMarker(
      location=[df.iloc[i]['Latitude_p'], df.iloc[i]['Longitude_p']],
      popup=[df.iloc[i]['cluster'], df.iloc[i]['Description']],
      radius=float(df.iloc[i]['Area_log']),
      color='#AA4A44',
      fill=True,
      fill_color='#AA4A44'
   ).add_to(new_m)
new_m
```
